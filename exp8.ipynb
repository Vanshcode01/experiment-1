```python
#q1a
with open("name.txt", "w") as f:
n = int(input("Enter number of names: "))
for i in range(n):
name = input()
f.write(name + "\n")
```

```python
#q1b
with open("name.txt", "r") as f:
names = f.readlines()
print("Total names:", len(names))
```

```python
#q1c
count = 0
with open("name.txt", "r") as f:
for line in f:
if line.strip()[0].lower() in "aeiou":
count += 1
print("Names starting with vowel:", count)
```

```python
#q1d
longest =""
with open("name.txt", "r") as f:
for line in f:
name = line.strip()
if len(name) > len(longest):
longest = name
print("Longest name:", longest)
```

```python
#q2a
with open("numbers.txt", "w") as f:
n = int(input("Enter count of numbers: "))
for i in range(n):
num = input()
f.write(num + "\n")
```

```python
#2b
with open("numbers.txt", "r") as f:
nums = [int(x.strip()) for x in f]
print("Maximum:", max(nums))
```

```python
#2c
count = 0
with open("numbers.txt", "r") as f:
numbers = [int(x.strip()) for x in f]
for num in numbers:
if num > 100:
count += 1
print("Numbers greater than 100:", count)
```

```python
#q4
n = int(input())
for i in range(n):
try:
a, b = input().split()
result = int(a) // int(b)
print(result)
except ZeroDivisionError as e:
    print("Error Code:", e)
except ValueError as e:
print("Error Code:", e)
```
